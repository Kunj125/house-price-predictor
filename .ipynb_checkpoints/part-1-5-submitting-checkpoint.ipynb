```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import tensorflow_decision_forests as tfdf
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
len(train)
```

```python

def convert_lot_area(row):
    if row['lotUnit'] == 'acres':
        return row['lotArea'] * 43560
    else:
        return row['lotArea']

def preprocess(df):
    df = df.drop(['address'], axis=1)
    df['lotArea'] = df.apply(convert_lot_area, axis=1)
    df = df.drop('lotUnit', axis=1)
    
    df['is_condo_or_townhouse'] = (df['homeType'] == 'CONDO') | (df['homeType'] == 'TOWNHOUSE')
    df.loc[df['is_condo_or_townhouse'], 'lotArea'] = 1
    df['is_condo_or_townhouse'] = df['is_condo_or_townhouse'].astype(int)
    return df
```

```python
train = preprocess(train)
test = preprocess(test)

# numerical_variables = ['bathrooms', 'bedrooms', 'livingArea', 'lotArea', 'price']
# train_no_outliers = train.copy()
# for variable in numerical_variables:
#     distance_from_mean = np.abs(train_no_outliers[variable].mean() - train_no_outliers[variable])
#     train_no_outliers = train_no_outliers[distance_from_mean < (train_no_outliers[variable].std() * 3)]

train_without_id = train.drop('id', axis = 1).copy()
train_without_id
```

```python
from datetime import datetime

train_without_id['dateSold'] = pd.to_datetime(train_without_id['dateSold'] / 1000, unit='s')
train_without_id['yearSold'] = train_without_id['dateSold'].dt.year
# train_without_id['bathroomsToBedroomsRatio'] = train_without_id['bathrooms'] / train_without_id['bedrooms']
train_without_id['bathroomsPlusrooms'] = train_without_id['bathrooms'] + train_without_id['bedrooms']
# train_without_id['livingAreaToLotAreaRatio'] = train_without_id['livingArea'] + train_without_id['lotArea']
# train_without_id = train_without_id.drop(columns=[ 'bedrooms', 'bathrooms'])

test['dateSold'] = pd.to_datetime(test['dateSold'] / 1000, unit='s')
test['yearSold'] = test['dateSold'].dt.year
# test['bathroomsToBedroomsRatio'] = test['bathrooms'] / test['bedrooms']
test['bathroomsPlusrooms'] = test['bathrooms'] + test['bedrooms']
# test['livingAreaToLotAreaRatio'] = test['livingArea'] + test['lotArea']
# test = test.drop(columns=['bedrooms', 'bathrooms'])

train_without_id = train_without_id.drop(columns=['dateSold', 'homeStatus'])
test = test.drop(columns=['dateSold', 'homeStatus'])

train_without_id
```

```python
len(test)
```

```python
def split_dataset(dataset, test_ratio=0.0001):
  test_indices = np.random.rand(len(dataset)) < test_ratio
  return dataset[~test_indices], dataset[test_indices]

train_ds_pd, valid_ds_pd = split_dataset(train_without_id)
print("{} examples in training, {} examples in testing.".format(
    len(train_ds_pd), len(valid_ds_pd)))
```

```python
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_ds_pd, label='price', task = tfdf.keras.Task.REGRESSION)
valid_ds = tfdf.keras.pd_dataframe_to_tf_dataset(valid_ds_pd, label='price', task = tfdf.keras.Task.REGRESSION)
```

```python
rf = tfdf.keras.RandomForestModel(task = tfdf.keras.Task.REGRESSION)
rf.compile(metrics=["mse"])
```

```python
rf.fit(x=train_ds)
```

```python
tfdf.model_plotter.plot_model_in_colab(rf, tree_idx=0)
```

```python
evaluation = rf.evaluate(x=valid_ds,return_dict=True)

for name, value in evaluation.items():
  print(f"{name}: {value:.4f}")
```

```python
ids = test.pop('id')
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(
    test,
    task = tfdf.keras.Task.REGRESSION)

pred = rf.predict(test_ds)
output = pd.DataFrame({'id': ids,
                       'price': pred.squeeze()})
output.head(20)
```

```python
submission = output
len(submission)
```

```python
submission.to_csv('submission.csv', index=False)
```
